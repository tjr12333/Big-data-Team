```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Untitled.csv", encoding='cp949')
df.head(10)
```

```python
data = df[['매물 구분','보증금','월세 금액','매매 금액']]

data.head(20)
```

```python
data2 = data[data["매물 구분"].str.contains("전세")].reset_index(drop=True)
data3 = data[data["매물 구분"].str.contains("월세")].reset_index(drop=True)
data4 = data[data["매물 구분"].str.contains("매매")].reset_index(drop=True)
data3.head(10)
```

```python
data2 = data2[['매물 구분','보증금']]
data3 = data3[['매물 구분','보증금','월세 금액']]
data4 = data4[['매물 구분','매매 금액']]
```

```python
data2_price=data2[['보증금']]
data3_price=data3[['보증금','월세 금액']]
data4_price=data4[['매매 금액']]
```

```python
data4_price
```

```python
from sklearn.preprocessing import MinMaxScaler
#정규화 진행
scaler2 = MinMaxScaler()
data_scale2 = scaler2.fit_transform(data2[['보증금']])

scaler3 = MinMaxScaler()
data_scale3 = scaler3.fit_transform(data3_price)

scaler4 = MinMaxScaler()
data_scal4e = scaler4.fit_transform(data4_price)
```

```python
from sklearn.cluster import KMeans
```

```python

# 데이터 스케일링
scaler3 = MinMaxScaler()
data_scale3 = scaler3.fit_transform(data3_price)

# 엘보우 방법을 사용하여 최적의 클러스터 수 찾기
distortions = []
K = range(1, 11)  # 클러스터 수 범위 설정

for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data_scale3)
    distortions.append(kmeans.inertia_)  # 응집도 저장

# 엘보우 그래프 그리기
plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Distortion')
plt.title('Elbow Method')
plt.show()
```

```python
# 최적의 클러스터 수로 K-means 클러스터링 수행
optimal_k = 3  # 최적의 클러스터 수 입력
kmeans = KMeans(n_clusters=optimal_k)
kmeans.fit(data_scale3)

# 클러스터링 결과 확인
labels = kmeans.labels_
```

```python
plt.scatter(data_scale3[:, 0], data_scale3[:, 1], c=labels)
plt.xlabel('보증금')
plt.ylabel('월세 금액')
plt.title('K-means Clustering Results')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)  # 원하는 클러스터 수에 따라 숫자를 변경하세요.
kmeans.fit(data_scale2)
```

```python
data2['Cluster'] = kmeans.labels_
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='보증금', y=data2.index, hue='Cluster', data=data2, palette='Set1')
plt.xlabel('보증금')
plt.ylabel('인덱스')
plt.title('k-means 클러스터링')
plt.show()
```

```python
k=2

# 그룹 수, random_state 설정
model = KMeans(n_clusters = k, random_state = 10)

# 정규화된 데이터에 학습
model.fit(data_scale2)

# 클러스터링 결과 각 데이터가 몇 번째 그룹에 속하는지 저장
df['cluster'] = model.fit_predict(data_scale2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (8, 8))

for i in range(k):
    plt.scatter(df.loc[df['cluster'] == i, 'Annual Income (k$)'], df.loc[df['cluster'] == i, 'Spending Score (1-100)'], 
                label = 'cluster ' + str(i))

plt.legend()
plt.title('K = %d results'%k , size = 15)
plt.xlabel('Annual Income', size = 12)
plt.ylabel('Spending Score', size = 12)
plt.show()
```
